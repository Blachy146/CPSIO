```python
"""
Wyświetlanie histogramu dla obrazu biało-czarnego
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)

plt.hist(image.ravel(), 256, [0, 256])
plt.xlim([0,256])

plt.show()
```

```python
"""
Wyświetlanie histogramu dla obrazu kolorowego
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

colors = ('b', 'g', 'r')

image = cv2.imread('./images/color_image.jpg')

for i, color in enumerate(colors):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color = color)
    plt.xlim([0, 256])
    
plt.show()
```

```python
"""
Wyrówanie histogramu dla obrazu czarno-białgeo
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)
equalized = cv2.equalizeHist(image)

plt.subplot(221), plt.imshow(image, 'gray')
plt.subplot(222), plt.imshow(equalized, 'gray')
plt.subplot(223), plt.hist(image.ravel(), 256, [0, 256])
plt.subplot(224), plt.hist(equalized.ravel(), 256, [0, 256])
plt.xlim([0,256])

plt.show()
```

```python
"""
Zastosowanie maski dla obrazu czarno-białego
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)

mask = np.zeros(img.shape[:2], np.uint8)
mask[70 : 180, 70 : 180] = 255

masked = cv2.bitwise_and(image, image, mask = mask)

image_hist = cv2.calcHist([image], [0], None, [256], [0, 256])
masked_hist = cv2.calcHist([image], [0], mask, [256], [0, 256])

plt.subplot(221), plt.imshow(image, 'gray')
plt.subplot(222), plt.imshow(mask, 'gray')
plt.subplot(223), plt.imshow(masked, 'gray')
plt.subplot(224), plt.plot(image_hist), plt.plot(masked_hist)
plt.xlim([0,256])

plt.show()
```

```python
"""
Szybka transformata Fouriera (FFT) dla obrazu czarno-białego
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)

fft_result = np.fft.fft2(image)
fft_result_shifted = np.fft.fftshift(fft_result)
magnitude_spectrum = 20 * np.log(np.abs(fft_result_shifted))

plt.subplot(221), plt.imshow(image, cmap = 'gray')
plt.subplot(222), plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.subplot(223), plt.hist(image.ravel(), 256, [0, 256])
plt.subplot(224), plt.hist(magnitude_spectrum.ravel(), 256, [0, 256])

plt.show()
```

```python
"""
Obraz po przepuszczeniu przez filtr górnoprzepustowy (wyodrębnia krawędzie)
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)

fft_result = np.fft.fft2(image)
fft_result_shifted = np.fft.fftshift(fft_result)
magnitude_spectrum = 20 * np.log(np.abs(fft_result_shifted))

rows, columns = image.shape
crow = int(rows / 2)
ccolumn = int(columns / 2)
fft_result_shifted[crow - 10 : crow + 10, ccolumn - 10 : ccolumn + 10] = 0
fft_ishift = np.fft.ifftshift(fft_result_shifted)
image_back = np.fft.ifft2(fft_ishift)
image_back = np.abs(image_back)

plt.subplot(221),plt.imshow(image, cmap = 'gray')
plt.title('Obraz oryginalny'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(image_back)
plt.title('FIltr górnoprzepustowy'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.hist(image.ravel(), 256, [0, 256])
plt.subplot(224), plt.hist(image_back.ravel(), 256, [0, 256])

plt.show()
```

```python
"""
Obrazy po zastosowaniu Laplasjana, SobelX i SobelY. Wyostrzanie
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)

laplacian = cv2.Laplacian(image, cv2.CV_64F)
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 5)

plt.subplot(2,2,1), plt.imshow(image, cmap = 'gray')
plt.title('Obraz oryginalny'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2), plt.imshow(laplacian, cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3), plt.imshow(sobelx, cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4), plt.imshow(sobely, cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
"""
Wysotrzanie krawędzi algorytmem Canny
"""

import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./images/black_white_image.jpeg', 0)
edges = cv2.Canny(image, 150, 200)

plt.subplot(121), plt.imshow(image, cmap = 'gray')
plt.title('Oryginał'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap = 'gray')
plt.title('Krawędzie'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
"""
Filtr Buterwortha dla sygnału EKG
"""

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

ekg = np.genfromtxt('./signals/ekg1', delimiter = '')

x = [0]
xi = [0]

for i in range(1, 8):
    x.append(x[i-1] + 1)
    xi.append(xi[i-1] + 800)

ekg_fft = np.fft.fft(ekg[:, 9])

N = 2 # Filter order
Wn = 0.01 # Częstotliwość graniczna
B, A = signal.butter(N, Wn, output = 'ba')
 
ekg_buter = signal.filtfilt(B, A, ekg[:, 9])
ekg_buter_fft = np.fft.fft(ekg_buter)

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.plot(ekg_buter)
plt.subplot(122), plt.plot(ekg_fft.real)

plt.show()
```
